```python
def visitor_ip_address(request):
    x_forwarded_for = request.META.get('HTTP_X_FORWARDED_FOR')
    if x_forwarded_for:
        ip = x_forwarded_for.split(',')[0]
    else:
        ip = request.META.get('REMOTE_ADDR')
    return ip
```

```python
import plotly.graph_objects as go


def show_chart(df,labels):
    
        # Use `hole` to create a donut-like pie chart
    fig = go.Figure(
        data=[go.Pie(labels=df['marital_status'], values=df['total'], hole=.3)])
    fig.update_traces(textinfo='label+percent+value',
                                    hoverinfo='label+percent', textfont_size=15)
    fig.update_layout(margin=dict(l=0, r=0, t=40, b=0),
                                    paper_bgcolor='rgba(0,0,0,0)',
                                    plot_bgcolor='rgba(0,0,0,0)',
                                    font=dict(size=15),
                                    showlegend=False,
                                    title="HTS_TST |"+ {labels}
                                    )
    # st.plotly_chart(fig, use_container_width=True)
    fig.show()

show_chart(cardinality('marital_status','education_level', 'sex','test_result'),labels)
```
